### 1.1 산업조직론의 정의와 목적


산업조직론은 기업 간의 경쟁, 시장 구조, 정부의 규제와 개입이 시장 성과에 미치는 영향을 분석하는 경제학의 한 분야로, 전통적인 미시경제학의 가격이론을 기반으로 발전해왔습니다. 이 이론은 완전경쟁과 독점이라는 두 극단 사이에 위치한 다양한 시장 형태, 예컨대 과점이나 독점적 경쟁 시장 등에서 기업들이 어떤 방식으로 행동하는지를 중점적으로 다룹니다. 주요 분석 대상에는 기업의 진입장벽 설정, 가격 결정 전략, 제품 차별화 방식, 연구개발(R&D) 투자 유인, 담합 가능성 등이 포함됩니다. 산업조직론은 이러한 기업 행위를 설명하고 예측하기 위해 게임이론, 계량경제모형, 사례연구 등의 다양한 도구를 활용합니다. 결국, 산업조직론은 시장에서 나타나는 기업의 행태와 그로부터 발생하는 성과를 체계적으로 분석함으로써 경쟁 메커니즘에 대한 이해를 높이고, 효과적인 산업 및 경쟁 정책 수립을 위한 시사점을 제공하는 데 목적이 있습니다.

**🎯 산업조직론의 목적**

**첫째, 시장 구조와 기업 전략의 상호작용 분석**

산업조직론의 가장 기본적인 목적은 과점이나 독점적 경쟁과 같은 불완전 경쟁 시장에서 기업이 어떻게 전략적으로 행동하는지를 분석하는 데 있습니다.
기업은 시장 상황에 따라 가격을 낮춰 경쟁하거나(가격 경쟁), 품질·브랜드·광고 등을 통해 차별화 전략을 취하기도 합니다(비가격 경쟁). 또한 진입장벽을 구축하여 신규 기업의 시장 진입을 억제하는 전략을 사용하기도 합니다. 산업조직론은 이러한 전략들이 시장 구조와 어떤 상호작용을 일으키는지를 체계적으로 설명합니다.

**둘째, 공정한 경쟁 환경 조성 및 규제 원칙 정립**

산업조직론은 정부 개입과 경쟁 정책의 이론적 정당성을 제공하는 데 중요한 역할을 합니다.
기업이 독점력을 남용하거나 담합하는 경우, 시장의 효율성과 소비자 후생이 저해될 수 있으므로, 반독점 정책과 경쟁법이 필요합니다. 산업조직론은 이러한 규제의 필요성과 타당성을 경제학적으로 설명하며, 시장지배력 규제, 담합 억제, 공정한 경쟁 촉진 등을 위한 정책 설계의 이론적 근거를 제시합니다.

**셋째, 시장성과의 평가**

시장 구조와 기업 행태가 실제로 어떤 **시장성과(market performance)**를 초래하는지를 평가하는 것도 산업조직론의 핵심 목적입니다.이를 위해 효율성(자원배분 효율성과 생산 효율성), 소비자 후생, 기업 수익성 등의 지표를 분석합니다. 전통적인 구조–행태–성과(SCP) 패러다임뿐 아니라, 정보 비대칭과 전략적 상호작용을 고려하는 **신산업조직론(NIO)**의 틀도 널리 활용됩니다.

**넷째, 산업 정책과 경쟁 정책 설계에 대한 이론적 기반 제공**

산업조직론은 정부가 수립하는 산업정책 및 경쟁정책의 효과를 예측하고 평가할 수 있는 분석도구를 제공합니다.예를 들어, 산업집중도가 높아질 경우 경쟁에 미치는 영향, 기업 간 M&A의 공정성 여부, 플랫폼 시장의 경쟁 구조와 같은 이슈에 대해 경제학적으로 접근하여, 정책 설계에 필요한 근거를 이론적으로 뒷받침합니다.

**다섯째, 디지털 및 글로벌 경제 환경에서의 경쟁 구조 해석**

최근 산업조직론은 플랫폼, 네트워크 효과, 데이터 독점 등 디지털 경제의 새로운 시장 구조를 해석하는 데로 분석 대상을 확장하고 있습니다.
또한 글로벌 가치사슬, 초국적 기업의 경쟁 전략과 같은 글로벌 경제 환경에서도 경쟁 구조를 이해하고 정책 대응을 논의하는 데 기여하고 있습니다. 이는 전통적인 이론을 넘어선 산업조직론의 응용 가능성을 보여주는 영역입니다.

**📌 산업조직론의 이론적 접근 방법**


산업조직론은 현실의 불완전 경쟁 시장을 설명하고 예측하기 위해 다양한 이론적 틀을 활용합니다. 가장 전통적인 분석 틀은 SCP 패러다임(Structure–Conduct–Performance)입니다. 이 접근은 시장구조(Structure)가 기업의 행태(Conduct)에 영향을 미치고, 그 결과로 시장성과(Performance)가 도출된다는 인과적 흐름을 전제로 합니다. 산업집중도, 진입장벽, 제품 차별화와 같은 구조적 특성이 기업의 전략적 선택에 영향을 준다고 보는 시각입니다.

하지만 현실의 경쟁은 단순한 구조적 분석만으로 설명되기 어려운 경우가 많습니다. 이에 따라 등장한 것이 게임이론 기반의 전략적 행태 분석입니다. 이 접근은 기업 간의 상호작용을 전략적 게임으로 모형화하여, 가격 담합, 진입 억제, R&D 투자 경쟁 등 다양한 행태를 분석합니다. 특히 과점 시장에서의 경쟁이나 담합은 반복게임이나 정보의 불완전성 등을 고려해야 하므로, 게임이론은 산업조직 분석에서 핵심적인 도구로 자리 잡고 있습니다.

이러한 흐름을 더욱 심화시킨 것이 신산업조직론(New Industrial Organization, NIO)입니다. 신산업조직론은 정보의 비대칭성, 유인 설계, 계약 구조 등 미시적 요소를 포함하여 기업 전략을 보다 현실적으로 설명합니다. 소비자 정보 부족, 기업 간 계약 조건, 잠재적 진입자의 존재 등 다양한 현실 제약을 고려함으로써, 전통 이론보다 한층 정교한 분석이 가능합니다.

### 1.2  미시경제학과 산업조직론의 관계

산업조직론은 미시경제학의 응용 분야로서, 가격이론·게임이론 등의 도구를 활용해 현실 시장의 기업 행태와 경쟁 구조를 심화 분석하고, 정책적 시사점을 도출하는 학문



**1. 이론적 기반: 미시경제학의 연장선** 

산업조직론은 **미시경제학, 특히 가격이론(Price Theory)**을 기반으로 발전한 하위 분야입니다.
미시경제학이 소비자와 생산자의 의사결정, 시장 균형, 자원 배분 등을 포괄적으로 다루는 반면, 산업조직론은 그중에서도 특히 기업 간 상호작용과 시장구조의 다양성에 주목합니다. 완전경쟁과 독점이라는 이론적 극단 외에도, 실제 경제에서 흔한 과점, 독점적 경쟁 시장의 복잡한 동학을 다루는 데 초점을 둡니다.

**2. 분석 범위의 확장** 

미시경제학은 주로 균형분석과 비교정태 분석을 통해 이론적 정합성을 추구합니다. 반면, 산업조직론은 전략적 상호작용, 게임이론, 정보 비대칭 등 미시이론의 심화 주제를 현실 시장에 적용하여 동태적이고 전략적인 기업 행동을 분석합니다. 예를 들어, 가격 설정이론은 미시경제학에서 기본적으로 다루는 주제이지만, 산업조직론에서는 가격 담합, 리더-추종자 모델, 진입억제 전략 등 구체적인 상황을 분석합니다.

**3. 응용 목적의 차별화** 

미시경제학은 이론 중심이며 경제 전체의 작동 원리에 대한 이해를 중시하는 반면, 산업조직론은 경쟁 정책, 독점 규제, 산업구조 개혁 등 정책적 함의 도출을 주요 목적으로 합니다. 따라서 산업조직론은 미시경제학의 분석 도구를 바탕으로 하되, 보다 현실 지향적인 응용 경제학의 성격을 강하게 띱니다.

## 2 산업조직론의 역사와 발전


산업조직론은 경제학의 실천적 영역으로서, 시장에서의 기업 행태와 경쟁 구조를 분석하는 목적을 가지고 발전해왔습니다. 이 분야는 20세기 초반의 초기 연구부터 게임이론 중심의 현대적 접근까지, 다음과 같은 주요 단계로 전개

---

### 1. 산업조직론 발전의 흐름

| 시기                         | 핵심 접근            | 특징적 내용                                                                                                                                                                     |
| -------------------------- | ---------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **1930‒1950s**<br>(초기 SCP) | **SCP 패러다임**     | • 산업집중도·진입장벽(구조) → 가격·광고 전략(행태) → 효율성·이윤·후생(성과) 단선 인과<br>• CR₄·CR₈ 집중률과 이윤 회귀분석, 독점사례 연구<br>• 집중도 높으면 담합 위험 ↑ → 사전 구조규제 정당화                                                |

| **1980s\~**<br>(신산업조직론)    | **게임이론·정보경제학**   | • 반복게임-기반 담합 지속, 신호이론·계약이론 도입<br>• 정·동태 게임·구조적 추정(Discrete Choice, GMM) 활용<br>• Merger Guidelines(1992·2010) 경쟁 시뮬레이션 근간<br>• ‘전략·인센티브’로 구조/행태 통합 분석                       |
| **2000s\~현재**<br>(디지털 IO)  | **플랫폼·네트워크·데이터** | • 양면시장 가격구조, 네트워크 효과·멀티호밍, 데이터 독점·알고리즘 담합<br>• 동태 점유율 전이·Churn, ML 기반 구조추정, 경매·메커니즘 디자인<br>• EU DMA(2024)·美 FTC 소송 ‒ 사전적(ex-ante) 플랫폼 규율<br>• 가격 외 ‘접근성·데이터 잠금’까지 경쟁 척도 확장 |


산업조직론의 전개는 대체로 네 가지 단계로 구분된다. 첫 단계인 1930~1950년대에는 하버드학파가 주도한 SCP 패러다임이 자리 잡았다. 이 접근은 산업집중도와 진입장벽 같은 구조적 요인에 따라 기업의 가격·광고 전략이 결정되고, 그 결과가 효율성·이윤·소비자 후생으로 귀결된다고 보았다. 연구자들은 CR₄·CR₈ 같은 집중률과 회계 이윤 사이의 통계를 교차-섹션 회귀로 분석하거나 독점 사례를 조사해, 집중도가 높을수록 담합 위험이 커지므로 사전적 구조 규제가 필요하다는 결론을 이끌어냈다.

1950~1970년대에 접어들면서 시카고학파가 등장해 기존 구조결정론을 정면으로 비판했다. 조지 스티글러와 해럴드 뎀세츠 등은 높은 집중도가 효율적 생산과 혁신의 결과일 수 있다고 주장하며, 경쟁 유지 여부를 판단하는 핵심 기준으로 시장 진입 용이성을 제시했다. 이들은 한계비용 가격결정, 거래비용·재산권 분석, 법경제학적 비용-편익 계산을 통해 정부 개입의 필요성을 재검토했고, 소비자 후생 중심의 탈규제 흐름—특히 1970년대 미국 항공·통신 자유화—을 이론적으로 뒷받침했다.

1980년대 이후 정보경제학과 게임이론의 확산, 컴퓨팅 능력의 발전을 배경으로 신산업조직론(New IO)이 부상했다. 장-티롤, 폴 밀그럼 등은 반복게임과 트리거 전략을 이용해 묵시적 담합의 지속 조건을 설명하고, 가격·품질을 통한 신호전달 모형, 수직계약과 다양화의 인센티브 구조를 제시했다. 구조적 추정(예: 이산선택모형, GMM)을 바탕으로 한 정·동태 게임 분석은 1992·2010년 미국 합병 가이드라인의 경쟁 시뮬레이션 모델로 이어졌고, 유럽연합 역시 제약·통신 분야의 수직담합 사건을 평가할 때 이러한 이론을 채택했다. 이 단계에서 연구 초점은 구조와 행태를 아우르는 전략·인센티브 분석으로 통합됐다.

2000년대 들어 디지털·플랫폼 경제가 확대되자 산업조직론은 플랫폼·네트워크·데이터 중심으로 확장된다. 로셰-티롤, 마크 암스트롱 등은 양면시장 이론을 통해 플랫폼의 가격 구조가 총수준보다 집단 간 배분에 의해 결정된다는 점을 규명했다. 네트워크 외부성과 멀티호밍, 데이터 독점, 알고리즘 담합이 주요 연구대상이 되었고, 동태 점유율 전이·Churn 모델, 머신러닝 기반 구조추정, 광고 경매·메커니즘 디자인이 분석 도구로 자리 잡았다. 정책 측면에서는 EU의 디지털시장법(DMA, 2024)과 미국 FTC·DoJ의 플랫폼 반독점 소송이 상징하듯, 가격 이외에도 접근성·데이터 잠금 같은 비가격 요소를 규제하는 사전적 감독 체계가 정착되고 있다.

이처럼 산업조직론은 구조결정론에서 효율성·진입 가능성, 전략·정보, 그리고 플랫폼·데이터 중심으로 분석 초점을 확장해 왔으며, 규제 역시 사전적 구조규제에서 탈규제, 행태·결합 규제, 사전적 플랫폼 규율로 단계적으로 진화해 왔다.

산업조직론의 초기 단계는 1930~1950년대에 형성되었으며, 당시의 시대적 맥락은 대공황 이후 미국 정부의 독점 규제 강화와 이에 대한 이론적 정당성 검토에 있었다. 특히, Clayton Act와 FTC Act와 같은 반독점 법제가 제정된 이후, 실제로 시장에서 독점이나 과점이 얼마나 존재하고, 그것이 경제성과에 어떤 영향을 미치는지에 대한 체계적 분석이 요구되었다. 또한 2차 세계대전 이후 산업 재건이 본격화되면서, 산업 구조에 대한 분석과 공정 경쟁의 토대 마련이 중요한 정책 과제로 떠올랐다.

이러한 배경에서 등장한 것이 하버드학파(Edward S. Mason, Joe S. Bain)의 분석 틀이며, 이들이 제시한 이론의 핵심은 바로 SCP 패러다임(Structure–Conduct–Performance)이다. 이는 시장의 구조(예: 산업집중도, 진입장벽)가 기업의 전략적 행태(예: 가격설정, 광고, 투자 등)를 결정하고, 그 결과로 나타나는 시장성과(예: 자원배분 효율성, 기업 이윤, 소비자 후생 등)를 설명하는 인과 사슬로 구성된다. 구조가 곧 행태를, 행태가 곧 성과를 설명한다는 단선적인 논리다.

이론의 실증적 검증을 위해 사용된 주요 분석 도구는 산업별 집중도 지표(CR₄, CR₈ 등)와 회계상 이윤율 간의 관계를 교차 섹션 회귀분석으로 추정하는 방식이었다. 또한 실제 독점 사례나 담합 사건에 대한 사례연구 방식도 널리 활용되었다. 이러한 분석은 단순히 학문적 의미를 넘어, 당시 미국 공정거래위원회(FTC)와 법무부(DoJ)가 독점 규제 사건을 심사하는 데 있어 직접적인 정책 프레임으로 채택되기도 했다.

하지만 SCP 패러다임은 곧 한계에 직면하게 된다. 구조 변수만으로 기업의 행태를 일방적으로 예측하는 것은 지나치게 단순화된 접근이라는 비판이 제기되었고, 특히 산업구조와 기업행태 간의 내생성 문제와 시장 동학의 무시는 현실을 설명하는 데 있어 중요한 약점으로 지적되었다. 이러한 비판은 이후 2단계 시카고학파의 효율성 가설과 시장자율주의적 접근의 이론적 기반이 된다.

결과적으로 초기 산업조직론은 산업 구조를 중심으로 시장의 경쟁성과를 평가하려는 정태적 분석 틀로서 정책·학문 양면에서 매우 중요한 출발점이 되었지만, 행태와 성과의 동태적 관계를 설명하기에는 이론적 유연성이 부족했다는 평가를 받는다.

시카고학파의 산업조직론: 효율성과 진입 가능성을 중심으로 한 전환기적 패러다임

1950~1970년대의 산업조직론은 하버드학파의 구조 중심적 사고에 대한 이론적 반작용으로서, 시카고학파(Chicago School)가 주도한 효율성 중심 접근법으로 전환되는 시기였다. 이 시대의 경제적 배경은 제2차 세계대전 이후의 전후 호황기와 규모의 경제 확장, 그리고 기존 규제에 대한 피로감이 누적되던 시기였다. 여기에 더해 프리드먼과 하이에크의 영향을 받은 신자유주의 경제사상이 학계와 정책에 본격적으로 유입되면서, 산업조직 분석에도 변화가 일기 시작했다.

이러한 변화의 중심에는 George Stigler, Harold Demsetz, Richard Posner 등이 있었으며, 이들은 SCP 패러다임이 제시한 '구조 → 행태'의 일방향 인과관계에 근본적 회의를 제기하였다. 이들은 시장의 집중도가 높다고 해서 반드시 시장지배력(market power)을 행사하거나 반경쟁적 행태가 나타나는 것은 아니라는 점을 강조하며, 오히려 집중도는 생산성 향상이나 혁신의 결과일 수 있다는 효율성 가설(효율성 가정)을 제안하였다. 즉, 집중도 상승은 시장지배력의 결과가 아닌, 비용 우위, 기술 혁신, 품질 경쟁에서의 승리에 따라 발생할 수 있는 자연스러운 시장 결과라는 것이다.

또한 시카고학파는 '진입 가능성(ease of entry)'을 경쟁의 실질적 기준으로 삼았다. 시장에 신규 기업이 자유롭게 진입할 수 있다면, 설령 소수의 기업이 시장을 점유하더라도 그 자체로는 문제 되지 않는다는 입장이다. 이는 이전 시기처럼 시장구조 자체를 규제하려는 노력보다, 실제 진입장벽이 존재하는지를 파악하고 규제의 필요성을 판단해야 한다는 주장으로 이어졌다.

분석 방법론 면에서도 이들은 한계비용定價(marginal cost pricing), 거래비용 이론(Coase, Williamson), 재산권 분석, 법경제학적 비용–편익 분석 등 미시경제학의 정합성과 논리성을 강조하며, 정성적·법적 사례보다는 수리적, 가격이론 중심의 분석을 중시했다. 특히 Richard Posner는 반독점 규제를 법경제학적 프레임으로 재해석하며, 소비자 후생을 유일한 규제 기준으로 제시하였고, 이는 로버트 보크(Robert Bork)의 『Antitrust Paradox』(1978)에서 정리되며 미국 경쟁 정책의 핵심 철학으로 자리 잡게 된다.

정책적 측면에서는 규제 최소화와 시장 자율 강조가 핵심 방향이었다. 1970년대 미국에서는 항공·통신·운송 산업 등 다양한 분야에서 자유화가 추진되었으며, 이는 시카고학파의 논리에 기반한 탈규제(deregulation) 흐름으로 설명될 수 있다. 시장구조에 대한 인위적 개입보다는 경쟁 자체가 시장 내에서 자연스럽게 작동하도록 하는 것이 바람직하다는 입장이 정책에 반영된 것이다.

이러한 시카고학파의 접근은 1단계의 SCP 패러다임이 가진 구조결정론적 한계에 대한 명확한 반응이자, 연역적 사고·미시경제학의 형식적 엄밀성을 산업조직론에 도입했다는 데 큰 의의가 있다. 다만 이 접근 역시 현실의 전략적 기업행태를 충분히 설명하지 못한다는 점에서, 이후 등장한 게임이론 기반의 신산업조직론(New IO)으로 이어지는 이론적 진화를 촉진하는 계기가 되었다.

신산업조직론(New IO)의 등장과 이론적 진화

1980년대 이후 산업조직론은 게임이론의 도입, 정보경제학의 발달, 그리고 계산능력의 비약적 향상을 계기로 질적 전환을 맞이하게 된다. 이 시기는 전통적 SCP 패러다임이 갖는 단선적 인과성과 정태성의 한계가 명확히 드러나던 시기였고, 시카고학파의 효율성 가설 또한 현실의 전략적 기업 행태(예: 묵시적 담합, 진입억제, 정보 비대칭 상황에서의 가격 전략 등)를 설명하기에 충분하지 않다는 평가가 확산되었다. 이에 따라 산업조직론은 기업 간 상호작용을 정교하게 분석할 수 있는 신산업조직론(New Industrial Organization, NIO) 체계로 진화하게 된다.

이론적 전환을 이끈 대표적인 학자는 Jean Tirole, Paul Milgrom, Bengt Holmström, Drew Fudenberg 등으로, 이들은 미시경제학의 심화 이론을 산업조직 분석에 체계적으로 접목시켰다. 특히 티롤(Jean Tirole)의 『The Theory of Industrial Organization(1988)』은 이후 산업조직론의 핵심 교과서로 자리잡았고, 정보의 비대칭성·계약이론·게임이론을 통합적으로 설명하는 틀을 제시했다.

신산업조직론은 기존과 달리 기업 간 전략적 상호작용을 중심으로 분석한다. 대표적인 개념은 반복게임과 Trigger Strategy를 통해 담합이 어떻게 장기적으로 유지될 수 있는지를 설명하는 것이다. 한두 차례의 게임이 아니라 무기한 반복되는 시장 상호작용에서 기업이 미래 이익 손실을 고려해 담합을 자발적으로 유지할 유인을 가진다는 점은, 과거 SCP 이론이 설명하지 못한 묵시적 담합을 이론화하는 데 기여했다.

또한 신호이론(signaling theory)은 정보 비대칭 상황에서 기업이 가격이나 품질을 통해 시장에 신호를 보내는 방식을 분석한다. 예컨대, 낮은 가격을 제시하는 것이 단순한 비용절감 때문이 아니라 진입자에게 “진입해봤자 손해”라는 시그널로 작용할 수 있다는 점은 진입억제 전략을 정교하게 이해하는 데 필수적이다. 이 외에도 다양화, 수직통합, 수직계약 등에서의 인센티브 구조 역시 정보와 계약 이론을 활용해 분석된다.

분석 방법론 면에서는 정태적·동태적 게임이론이 광범위하게 사용된다. 베이지안 게임, 시그널링 게임, 스크리닝 모델 등은 현실의 기업 전략을 수학적으로 모형화하고 예측하는 데 유용하며, 이를 실증적으로 검증하기 위해 구조적 추정 방법(structural estimation)도 함께 발전했다. 예컨대 Discrete Choice 모델과 GMM(Generalized Method of Moments)을 활용한 수요·가격 결정 구조 추정은 정책 효과 분석에도 직접적으로 활용된다. 또한 계약이론(contract theory)은 정보의 비대칭 하에서 기업 간 계약 조건이 어떻게 설정되고 작동하는지를 설명하는 이론으로, 수직통합·라이선싱·프랜차이즈 등 다양한 산업적 맥락에서 적용된다.

정책적으로는 신산업조직론이 미국과 유럽의 경쟁정책에 광범위하게 적용되기 시작했다. 특히 미국 법무부(DoJ)의 1992년 및 2010년 Merger Guidelines에서는 구조적 집중도 지표뿐만 아니라, 시장 수요탄력성·가격 반응 함수 등을 이용한 경쟁 시뮬레이션 기법이 도입되었다. 이는 신산업조직론에서 제시한 모형과 방법론이 직접적으로 정책 결정에 활용되기 시작했음을 보여주는 사례이다. 유럽연합 집행위원회(European Commission) 또한 제약, 통신, 플랫폼 산업에서의 장기 계약·수직제약 사건을 분석할 때 NIO 접근을 적극 활용하였다.

요약하자면, 신산업조직론은 기존의 “구조 → 행태” 또는 “효율성 중심 분석”이라는 단편적 구도에서 벗어나, 기업의 전략적 판단과 유인 구조를 분석 중심에 놓는 통합적 접근으로 발전했다. 이는 이전 단계들과 비교해 현실 적합성, 수리적 정밀성, 정책 응용력이 모두 강화된 전환점이며, 오늘날 산업조직 분석의 기본 틀로 자리잡고 있다.

2000년대 이후 산업조직론은 다시 한번 큰 전환점을 맞이하게 된다. 인터넷과 스마트폰의 확산, AI·클라우드 기반의 데이터 자본 축적, 네트워크 외부성의 강화 등으로 대표되는 디지털 경제의 부상은 기존의 경쟁 분석 프레임이 갖는 한계를 드러냈다. 특히 소수의 거대 플랫폼 기업들이 시장접근, 데이터통제, 네트워크 구축을 통해 시장지배력을 강화하는 현실은 전통적인 가격 중심 분석이나 수직결합 모형만으로는 설명하기 어려운 구조적 문제들을 동반했다.

이러한 배경에서 주목받은 이론이 바로 양면시장(two-sided markets) 이론이며, 이를 정립한 대표 학자들이 Jean-Charles Rochet & Jean Tirole, Mark Armstrong, David Evans 등이다. 이들은 시장에서 플랫폼이 양쪽 사용자 그룹(예: 소비자와 앱 개발자, 광고주와 이용자 등)을 연결하는 중개자로서 작동할 경우, 가격 구조(price structure)가 단일 상품 시장과 전혀 다르게 형성될 수 있음을 보였다. 예컨대 한쪽 집단(소비자)에게는 무료 서비스를 제공하면서 다른 집단(광고주 등)에게 수익을 집중시킨다는 식이다. 여기서 핵심은 총 가격(price level)이 아니라 가격 배분(price allocation)이며, 이는 전통 산업조직론과 명확히 구분되는 지점이다.

이와 더불어 네트워크 효과(network effects)와 멀티호밍(multi-homing) 문제도 플랫폼 경제의 핵심 분석 대상이 되었다. 네트워크 효과는 가입자가 늘어날수록 가치가 커지는 구조이며, 이는 소수 기업의 독점적 지위를 강화시키는 메커니즘으로 작동한다. 멀티호밍은 사용자가 여러 플랫폼을 동시에 사용하는 전략이지만, 이를 제한하기 위한 잠금효과(lock-in)와 데이터 전환 비용 역시 새로운 진입장벽으로 작용한다. 여기에 최근에는 데이터 독점, 알고리즘 기반 가격 조정 및 담합, 선별적 가시성(디폴트 설정, 검색 우선순위) 등이 주요 쟁점으로 떠올랐다.

분석 방법론도 진화했다. 정적 게임이론 분석이나 구조추정 모형에서 벗어나, 플랫폼 산업에서는 동태적 점유율 전이 모델, 사용자 이탈(Churn) 예측, 대규모 로그데이터 기반의 머신러닝 구조추정이 결합되고 있다. 또한 광고 경매나 디지털 클라우드 입찰 구조 분석을 위해 경매이론(auction theory)과 메커니즘 디자인(mechanism design)이 핵심 도구로 쓰이고 있다. 예를 들어 구글의 검색광고 입찰 구조는 산업조직론+경매이론의 융합 사례다.

정책적 시사점 측면에서는 규제 철학의 대전환이 이뤄졌다. 전통적 경쟁정책은 시장지배력 남용을 사후적(ex-post)으로 판단했으나, 디지털 플랫폼은 ‘접근 자체의 통제’, ‘초기 진입 봉쇄’, ‘데이터 수직 통합’ 등 사전적 위험이 크기 때문에, 규제도 사전적(ex-ante) 접근이 요구되기 시작했다. 유럽연합(EU)은 2024년부터 디지털시장법(DMA)을 발효시켜, 일정 기준 이상의 플랫폼 기업(게이트키퍼)에 대해 타사 앱 접근 허용, 데이터 혼용 금지, 자사 우대 금지 등의 규제를 사전에 부과하도록 했다. 미국 역시 FTC와 DoJ가 구글·애플에 대한 반독점 소송을 통해, 검색 알고리즘 조작, 앱스토어 수수료, 플랫폼 내 차별적 조건 등을 담합 또는 시장봉쇄로 간주하고 적극 대응에 나섰다.

이러한 변화는 기존 산업조직론이 가격 중심의 단면적 분석에 머물렀던 것에서 벗어나, 디지털 자산의 다면적 가치(value-multisidedness), 비가격적 경쟁 요소(접근성, 호환성, 가시성), 데이터 축적의 전략성 등을 중심으로 분석 대상과 개념틀을 확장하고 있음을 보여준다. 예컨대, 애플 vs 에픽 게임즈 소송은 단순히 수수료 수준의 문제가 아니라, 앱스토어 접근권, OS-API의 봉쇄, 자체 결제 차단 등 양면시장 이론과 수직제약 이론이 동시에 적용되는 복합적 구조를 보여주는 대표 사례다.

| **3. 전략·정보론**<br>(1980s\~, 신산업조직론)    | 게임·정보경제학으로 **전략적 상호작용** 분석           | 반복게임 담합·신호이론·계약인센티브        | 1990s **BA–버진 로열티 스킴** 사건 → 트리거 전략·신호이론 적용         |
| **4. 네트워크·데이터론**<br>(2000s\~, 디지털 IO) | 플랫폼 **양면시장·네트워크 효과·데이터 잠금** 중심       | 가격 구조보다 집단 간 **배분**·접근성 규제 | 2020s **애플 앱스토어** 수수료·API 봉쇄 → 양면시장‧수직제약·데이터 독점 이슈 |


산업조직론의 각 발전 단계는 현실 사례를 통해 이론적 타당성을 시험받아 왔습니다. 아래에서는 대표 사례와 그 사례가 어떻게 해당 단계의 핵심 이론을 검증했는지를 서술형으로 정리했습니다.

1단계 (SCP 패러다임) – 미국 철강·석유 집중도 조사 (1940년대)
1940년대 미국 연방거래위원회(FTC)는 철강‧석유 산업의 산업집중도(CR₄·CR₈)와 기업 이윤 간 관계를 대규모로 조사했다. 분석 결과, 집중도가 높은 시장일수록 회계상 초과이윤이 유의하게 높게 나타났고, 이는 ‘구조(집중도) → 행태(가격 인상 및 담합 유인) → 성과(이윤·후생 손실)’라는 SCP 패러다임의 단선 인과를 실증적으로 뒷받침했다. 이 연구는 “집중도 상승 = 담합 위험 증가”라는 정책적 메시지로 이어져, 당시 독점 금지 행정의 사전 구조규제 논거를 마련했다.

2단계 (시카고학파) – 시카고 화물터미널 규제 완화 (1970년대)
1970년대 미국 정부는 시카고 일대 화물터미널에 대한 진입 제한과 요금 규제를 완화했다. 규제 철폐 후 다수의 신규 운송업체가 시장에 진입했으며, 가격 경쟁이 심화되면서 화물 운임이 하락하고 서비스 품질이 향상되었다. 이 사례는 “경쟁 유지의 핵심은 시장 구조가 아니라 ‘진입 가능성’”이라는 시카고학파 주장을 실제로 입증했다. 높은 집중도 자체를 규제하기보다 진입 장벽을 낮추는 탈규제 정책이 더 효과적일 수 있음을 보여준 것이다.

3단계 (신산업조직론) – 브리티시항공 vs 버진 어틀랜틱 담합 판결 (1990년대)
1990년대 영국 공정거래청은 브리티시항공(BA)이 버진 어틀랜틱(VA)과 체결한 로열티 프로그램(Travel Incentive Scheme)을 담합으로 판단했다. 조사는 반복게임 기반 ‘트리거 전략’과 신호이론으로 분석되었는데, BA가 가격을 직접 인하하지 않고도 로열티 스킴을 통해 VA의 가격 인하 시도를 억제한다는 전략적 신호를 시장에 보낸 것으로 해석됐다. 이는 신산업조직론이 강조하는 전략적 상호작용과 정보 비대칭이 실제 담합 유지 메커니즘으로 작동함을 보여준 대표적 판례다.

4단계 (디지털 IO) – 애플 앱스토어 수수료·타이핑 API 사건 (2020년대)
2020년대 애플은 자사 앱스토어(양면시장)에서 인앱 결제 수수료(30%)를 부과하고, 서드파티 키보드 앱이 시스템 타이핑 API에 접근하지 못하도록 제한해 왔다. 에픽게임즈 소송 및 각국 규제 당국 조사에서 핵심 쟁점은 수수료 구조가 개발자·소비자 양 측면을 어떻게 배분하는지(양면시장 이론)와, OS·API 봉쇄가 데이터 잠금·수직제약을 통해 진입을 억제하는지였다. 이 사례는 플랫폼 시대 산업조직론이 강조하는 데이터 독점, 접근성 규제, 비가격 경쟁 요소가 실제 시장지배력 행사 수단이 될 수 있음을 실증적으로 보여준다.

위 네 사례는 각각의 시대가 직면한 시장 현실과 정책 과제를 반영하며, 산업조직론 이론이 어떻게 현실 검증과 정책 설계에 활용되어 왔는지를 구체적으로 드러낸다.

---

산업조직론의 분석 초점은 시대 흐름에 따라 네 차례 큰 전환을 겪었다. 가장 먼저 구조결정론 시기에는 산업집중도나 진입장벽 같은 구조적 변수로 기업 행태와 시장 성과를 일방적으로 설명하려 했다. 이어 효율성·소비자후생 중심 단계에 들어서면서, 높은 집중도가 반드시 독점 폐해를 의미하지 않을 수 있으며, 진입 용이성과 자원의 효율적 배분이 경쟁 여부를 가른다는 관점이 자리 잡았다. 1980년대부터는 게임이론과 정보경제학이 도입되며 전략·정보 중심 분석으로 무게가 이동했고, 2000년대 이후 디지털 플랫폼이 등장하면서 네트워크·데이터 중심 연구가 부상했다. 이제는 네트워크 효과, 멀티호밍, 데이터 잠금 같은 요인이 경쟁력을 결정짓는 핵심 요소가 되고 있다.

정책 패러다임도 이에 맞춰 진화했다. 초기에는 구조적 집중도를 기준으로 사전적으로 산업을 규제했으나, 곧 효율성과 소비자 후생을 중시하는 탈규제 흐름이 확산됐다. 이후 전략적 담합과 혁신 유인을 동시에 고려하는 행태·결합 규제로 무게중심이 이동했고, 오늘날에는 플랫폼의 시장력과 데이터 지배력을 사전에 감독하는 디지털 거버넌스 체계가 자리 잡고 있다. 즉, 가격이라는 단일 척도를 넘어 접근성, API 개방성, 데이터 이동권 등 비가격 요소가 정책의 핵심 잣대로 떠올랐다.

연구 방법론 또한 시대별 데이터 특성과 산업 환경에 따라 달라진다. 구조결정론기에는 산업별 집중률과 이윤을 단순 회귀로 분석하면 충분했지만, 효율성·진입 단계에서는 거래비용 계산과 비용–편익(B/C) 비교가 중요했다. 전략·정보기가 도래하자 수요·가격·행태를 세밀하게 추적할 수 있는 패널 자료와 함께 게임이론 기반의 구조적 추정(Discrete-choice, GMM)이 표준도구로 자리 잡았다. 디지털 시대에는 대규모 플랫폼 로그, API 호출 기록 등 빅데이터를 활용한 머신러닝·경매 분석이 필요하며, 네트워크 메트릭을 동태적으로 추정하는 기법이 필수다. 결국 연구자는 사용 가능한 데이터의 범위, 산업의 특수성, 그리고 정책 목표를 종합적으로 고려해 적절한 방법론을 택해야만 실증 결과와 정책 처방 사이의 일관성을 보장할 수 있다.

산업조직론(기업과 시장을 연구하는 경제학 분야)은 처음에 **‘구조가 모든 걸 결정한다’**는 생각에서 출발했습니다. 1930~1950년대 하버드대의 에드워드 메이슨과 조 베인 같은 학자들은 SCP(Structure–Conduct–Performance) 모형을 내세웠는데, “산업에 몇 개 거대기업이 버티고 있느냐(집중도)와 새 기업이 들어오기 어렵냐(진입장벽)가 높으면, 기업들은 서로 짜고 가격을 올리고 결국 소비자가 손해 본다”는 주장이었죠. 실제로 이들은 미국 철강·석유 산업을 조사해 집중도가 높을수록 이윤이 높았다는 통계를 내놓았고, 이 결과는 “큰 회사는 나눠 놓자”는 분할 명령 같은 사전적 구조 규제의 근거가 됐습니다.

하지만 시간이 지나자 “구조만 보고는 현실을 설명하기엔 너무 단순하다”는 비판이 쏟아졌습니다. 시장은 움직이고(동태적), 구조와 행태가 서로 영향을 주고받기 때문입니다. 이를 보완하려고 1980년대부터 등장한 것이 신구조주의(Structural IO) 입니다. 연구자들은 컴퓨터 계산 능력과 게임이론·정보경제학을 활용해 소비자 취향, 생산비용, 기업 전략을 수식으로 자세히 추정하고, 두 회사가 합치면 가격과 소비자 후생이 어떻게 변할지를 가상 실험(합병 시뮬레이션) 으로 예측했습니다. 미국 경쟁당국은 1992·2010년 합병 심사 지침에서 단순 점유율 대신 이런 구조추정 결과를 공식 채택하여, 사건마다 ‘가격이 몇 % 오를지’까지 정밀하게 따지게 되었죠.

한편, 게임이론은 “기업끼리 서로를 의식하며 머리싸움을 한다”는 점에 집중합니다. 치킨집 두 곳이 가격을 내릴지 말지 고민하는 상황을 경기(게임)로 그리고, 여러 번 반복되면 “오늘은 올리지 말고 너도 올리지 마” 같은 묵시적 담합이 형성될 수 있음을 보여줍니다. 이런 모델 덕분에 정부는 “값이 이상하게 비슷하다”는 증거만으로도 담합 조사에 착수할 수 있습니다.

거래비용 이론은 “직접 만들까, 사올까?”를 따집니다. 햄버거 가게가 빵을 외주로 살지 매장 뒤에서 직접 구울지 결정할 때, 자산특수성(나만 쓰는 오븐), 거래 빈도, 가격 변동 위험 등을 모두 비용으로 계산합니다. 이 관점은 영화사가 영화관까지 인수하는 수직합병이 효율 때문인지, 경쟁자를 막으려는 꼼수인지 판단하는 기준이 됩니다.

마지막으로 경쟁과 협력(Co-opetition) 이론은 “기업이 때로는 싸우고, 때로는 손잡는다”는 현실을 설명합니다. 예컨대 스마트폰 회사들이 충전 단자를 통일하면 시장이 커지지만, 동시에 가격 담합 우려도 생깁니다. 그래서 표준 필수특허(SEP)는 누구나 쓰게 하되 공정한 사용료만 받도록 하고, 앱스토어처럼 폐쇄적인 플랫폼에는 일정 수준의 데이터·API 개방을 요구하기도 합니다.

종합하면, 구조주의는 시장 규제의 출발점을 마련했지만 과잉 규제 위험이 있었고, 신구조주의·게임이론·거래비용·협력 이론은 복잡한 현대 시장을 이해하기 위해 데이터·전략·조직·네트워크까지 통합적으로 살펴봅니다. 연구자와 정책 담당자는 사용 가능한 데이터의 양과 질, 산업의 특수성, 그리고 “가격 안정이 목표인가, 혁신 촉진이 목표인가?” 같은 정책 목표를 함께 고려해, 회귀나 비용-편익 분석부터 구조적 추정, 머신러닝·경매 모델까지 적절히 조합해야 합니다. 그래야 연구 결과와 규제 처방이 서로 모순되지 않고, 소비자와 기업 모두에게 이로운 결정을 내릴 수 있습니다.
