# 산업조직론 기초

**학습 요약**

이 목차는 시장 구조 분석과 경쟁 이론을 시작으로, 기업의 전략적 행동, 정보 비대칭 문제, 수직적 통합을 다룹니다. 마지막으로 산업조직에 대한 정부 규제와 글로벌화까지 포괄적으로 학습할 수 있습니다. 이를 통해 시장 구조와 기업 간 경쟁의 메커니즘을 심도 있게 이해하고, 다양한 시장 상황에서의 경쟁 전략과 정책적 대응을 분석하는 데 유용합니다.

### 산업조직론 학습 목차

---

### 1. **산업조직론의 개요**

- **1.1 산업조직론의 정의와 목적**
    - 산업조직론의 연구 범위와 목적
    - 미시경제학과 산업조직론의 관계
- **1.2 산업조직론의 역사와 발전**
    - 초기 산업조직 연구부터 현대 이론의 발전 과정
    - SCP 패러다임(구조-행위-성과 모형)의 등장과 한계
- **1.3 산업조직론의 주요 연구 접근법**
    - 구조주의적 접근과 신구조주의적 접근
    - 게임이론, 거래비용 이론, 경쟁과 협력 이론

### 2. **시장 구조와 기업 간 경쟁**

- **2.1 시장 구조의 분류**
    - 완전 경쟁, 독점적 경쟁, 과점, 독점의 개념과 특징
    - 각 시장 구조에서 기업의 전략적 행동
- **2.2 집중도와 시장 파워**
    - 집중도 지표 (CR4, HHI 등)와 시장 지배력
    - 시장 파워의 측정 방법과 의미
- **2.3 진입 장벽과 퇴출 장벽**
    - 진입 장벽의 유형(규모의 경제, 절대 비용 우위 등)
    - 퇴출 장벽의 존재와 산업 내 경쟁성에 미치는 영향
- **2.4 기업 간 상호작용과 시장 구조**
    - 시장 구조와 가격 결정의 상호작용
    - 기업 간 전략적 상호작용이 시장에 미치는 영향

### 3. **독점과 과점 이론**

- **3.1 독점 시장과 기업의 가격 결정**
    - 독점기업의 가격 및 생산량 결정
    - 소비자 후생과 독점의 경제적 비효율성
- **3.2 자연 독점과 규제**
    - 자연 독점의 개념과 발생 원인
    - 정부 규제의 필요성과 공공요금 설정
- **3.3 과점 시장의 경쟁과 협력**
    - 과점 시장의 특징과 기업 간 상호작용
    - 가격 리더십, 담합, 비가격 경쟁 전략
- **3.4 과점 시장의 게임 이론적 접근**
    - 게임 이론과 과점에서의 전략적 상호작용
    - 죄수의 딜레마, 내쉬 균형, 반복 게임

### 4. **경쟁 전략과 기업 행동**

- **4.1 가격 전략**
    - 가격 차별화 전략(1급, 2급, 3급 가격 차별화)
    - 약탈적 가격 책정과 시장 파괴 전략
- **4.2 제품 차별화와 비가격 경쟁**
    - 제품 차별화의 개념과 경제적 효과
    - 광고, 브랜드 이미지, 품질 개선을 통한 차별화 전략
- **4.3 혁신과 연구개발(R&D)**
    - 혁신과 R&D가 경쟁 우위에 미치는 영향
    - 특허와 지적 재산권의 역할
- **4.4 시장 진입과 진입 방어 전략**
    - 진입 방어 전략(선발적 진입 장벽 구축, 용량 과잉)
    - 신생 기업의 진입 전략과 기존 기업의 방어 전략

### 5. **게임 이론과 산업조직**

- **5.1 게임 이론의 기초 개념**
    - 기본 개념(전략, 보수행렬, 내쉬 균형)
    - 내쉬 균형의 의미와 경제적 적용 사례
- **5.2 반복 게임과 협력 가능성**
    - 반복 게임과 협력 유지 조건
    - 무한 반복 게임과 민감한 협력 전략 분석
- **5.3 혼합 전략 균형**
    - 혼합 전략 균형의 개념과 적용
    - 확률적 선택과 불확실성 하에서의 전략적 행동
- **5.4 경로 의존성 및 신호 게임**
    - 경로 의존성의 의미와 기업의 전략적 선택
    - 신호 전달 게임을 통한 정보 비대칭 해결

### 6. **정보 비대칭과 시장 실패**

- **6.1 정보 비대칭의 개념과 문제점**
    - 정보 비대칭이 시장에 미치는 영향과 사례
- **6.2 주인-대리 문제와 도덕적 해이**
    - 주인-대리 문제, 도덕적 해이 및 역선택 문제
    - 기업 내 주인-대리 문제와 대리인 비용
- **6.3 신호와 선별 기제**
    - 신호 이론을 통한 정보 비대칭 완화 방법
    - 선별 기제를 통한 시장 효율성 향상

### 7. **기업의 수직적 관계와 통합**

- **7.1 수직적 통합의 경제적 효과**
    - 수직적 통합(상향·하향 통합)의 경제적 효과
    - 수직적 통합과 비용 절감, 시장 파워 강화
- **7.2 수직적 계약과 거래비용 이론**
    - 거래비용 이론과 기업 경계의 경제적 의미
    - 장기 계약, 독점 공급 계약 등의 전략적 활용
- **7.3 프랜차이즈와 전략적 제휴**
    - 프랜차이즈와 전략적 제휴의 경제적 효과
    - 브랜드 확장과 통합 전략의 장단점

### 8. **규제와 공공정책**

- **8.1 독점 규제와 반독점법**
    - 독점 규제의 필요성과 반독점법의 주요 조항
    - 시장지배적 지위 남용 방지 및 공정 경쟁 촉진
- **8.2 가격 규제와 공공요금 결정**
    - 가격 상한제와 공공요금 결정 방식
    - 사회 후생을 위한 가격 규제의 경제적 효과
- **8.3 외부효과와 환경 규제**
    - 외부효과와 시장 실패 문제
    - 환경 규제를 통한 외부비용 내부화
- **8.4 혁신 촉진과 정부 지원**
    - 연구개발 지원 정책과 혁신 장려의 경제적 효과
    - 특허제도와 혁신 촉진을 위한 정부의 역할

### 9. **산업 조직과 국제 경쟁**

- **9.1 글로벌화와 산업조직**
    - 글로벌화가 산업 구조와 기업 경쟁에 미치는 영향
    - 글로벌화에 따른 기업의 경쟁 전략 변화
- **9.2 국제 시장에서의 기업 전략**
    - 다국적 기업의 진출과 시장 점유 전략
    - 해외 시장에서의 진입 장벽과 현지화 전략
- **9.3 무역과 산업 정책**
    - 무역 정책이 산업조직에 미치는 경제적 영향
    - 보호주의와 자유무역의 산업별 효과 비교

### 10. **산업조직론의 최신 연구 동향**

- **10.1 행동경제학과 산업조직**
    - 비합리적 소비자 행동이 산업조직에 미치는 영향
    - 제한된 합리성과 시장 구조
- **10.2 디지털 경제와 플랫폼 시장**
    - 디지털화가 기업 간 경쟁 및 시장 구조에 미치는 영향
    - 플랫폼 경제의 특성 및 네트워크 효과
- **10.3 기술 혁신과 산업 변화**
    - 기술 혁신이 시장 구조에 미치는 영향
    - 자동화와 인공지능의 도입으로 인한 산업 재편성
- **10.4 산업조직론의 미래 과제**
    - 공유 경제, 인공지능 등 미래 경제 환경에서의 산업조직론 과제
    - 지속 가능한 산업 구조를 위한 경제적 접근

# 산업조직론의 심화 학습 주제

---

### 1. **고급 시장 구조 분석**

- **1.1 다각적 시장 구조와 동태적 변화**
    - 시장 구조의 동적 변화: 시장 진입과 퇴출이 장기적으로 미치는 영향
    - 시장 집중도와 그 변화의 경제적 효과 분석
- **1.2 다중 시장 접촉(Multi-market Contact)**
    - 다중 시장에서의 경쟁과 협력의 역할
    - 항공, 통신 등 다중 시장 접촉이 빈번한 산업에서의 전략적 행위 분석
- **1.3 진입과 퇴출 전략의 동학**
    - 진입 및 퇴출 장벽이 시장 성과에 미치는 장기적 효과
    - 신생 기업의 진입 전략과 기존 기업의 방어 전략 사례 연구

### 2. **경쟁 전략의 고급 분석**

- **2.1 동태적 가격 결정과 약탈적 가격 책정**
    - 약탈적 가격 책정의 전략적 사용과 규제 대응
    - 동태적 가격 조정 전략(예: 가격 경기 관찰과 반응)
- **2.2 합병 및 인수 (M&A)와 기업 성장**
    - M&A가 시장 구조와 경쟁에 미치는 경제적 영향
    - 인수합병에 대한 규제와 반독점법의 대응
- **2.3 제품 차별화와 경쟁 우위 지속성**
    - 제품 차별화 전략의 심화 분석과 브랜드 유지 전략
    - 디자인, 품질, 서비스 등 비가격 요소가 경쟁력에 미치는 장기적 효과

### 3. **게임 이론을 통한 전략적 상호작용**

- **3.1 동태적 게임과 반복 게임**
    - 반복 게임의 협력 및 배신 전략 분석
    - 무한 반복 게임에서의 민감한 협력 전략과 적용 사례
- **3.2 혼합 전략과 정보 비대칭 하의 경쟁**
    - 불완전 정보 하에서의 전략적 선택과 혼합 전략
    - 불확실성과 정보 비대칭이 경쟁에 미치는 영향
- **3.3 경로 의존성과 진화적 게임**
    - 경로 의존성과 누적적 경쟁 우위 형성
    - 진화적 게임 이론을 통한 기업 간 장기적 경쟁 구도 분석

### 4. **독점적 시장과 독점력 유지 전략**

- **4.1 독점적 가격 책정과 시장 지배력**
    - 독점적 가격 책정 전략(가격 차별화, 분리 정가제)
    - 독점기업의 가격 책정과 후생경제학적 영향
- **4.2 규제 회피와 독점력 유지**
    - 독점적 기업이 규제를 회피하거나 완화하는 전략
    - 외형적 경쟁 유지(약탈적 가격 책정, 과잉 설비 구축 등)로 인한 시장 지배력 유지
- **4.3 기업의 시장 파워와 반독점 규제 대응 전략**
    - 시장 파워를 이용한 기업의 전략적 조치와 규제 회피
    - 반독점법 적용의 사례와 그 경제적 효과

### 5. **정보 비대칭과 시장 실패의 고급 분석**

- **5.1 정보 비대칭과 신호 전달 이론의 응용**
    - 비대칭 정보 하에서의 신호 전달 이론과 시장 효율성 개선
    - 금융 시장에서의 정보 비대칭과 신용 할당
- **5.2 선별과 스크리닝 기법**
    - 정보 비대칭 하에서 선별 기법(보험, 금융, 노동 시장 등) 적용
    - 스크리닝 전략을 통해 시장 효율성을 높이는 방법
- **5.3 계약 이론과 주인-대리 문제 심화**
    - 주인-대리 문제에 대한 계약 이론 적용 및 경제적 효과
    - 대리인 비용을 줄이기 위한 최적 계약 설계

### 6. **혁신과 기술 변화가 산업조직에 미치는 영향**

- **6.1 혁신과 시장 구조의 동태적 변화**
    - 기술 혁신이 기존 시장 구조에 미치는 영향과 재편성 과정
    - 혁신 주도형 산업에서 경쟁력 유지 방법
- **6.2 R&D와 특허 전략**
    - R&D 투자의 경제적 분석과 기업 간 협력 전략
    - 특허 제도를 통한 경쟁 우위와 정부의 혁신 촉진 정책
- **6.3 기술 주기 이론과 산업 구조의 재편**
    - 기술 주기 이론을 통해 본 산업 재편성 과정
    - 기술 혁신의 속도와 기업 생존 가능성 간의 관계 분석

### 7. **디지털 경제와 플랫폼 산업조직**

- **7.1 플랫폼 경제의 특징과 네트워크 효과**
    - 플랫폼 경제에서의 네트워크 외부성과 경쟁 구조 분석
    - 직접 네트워크 효과와 간접 네트워크 효과의 경제적 의미
- **7.2 다면시장(Multi-sided Market)과 가격 전략**
    - 다면시장의 구조와 가격 전략의 경제적 분석
    - 플랫폼 간 경쟁과 보조금 전략
- **7.3 데이터와 정보의 경제적 가치**
    - 플랫폼 시장에서 데이터의 경제적 가치와 독점력
    - 데이터 독점이 산업조직에 미치는 장기적 영향

### 8. **규제와 산업정책의 심화 분석**

- **8.1 규제와 규제 포획 현상**
    - 규제 포획 이론과 기업이 규제를 활용하는 방법
    - 규제기관과 기업 간 상호작용이 경제에 미치는 효과
- **8.2 독점 규제와 공공정책**
